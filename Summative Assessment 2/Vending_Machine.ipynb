```python
# The variables the programme requires.
credits = 1075
entered_amount = 0 
kept_amount = 0
overall_cost = 0
snack_count = 0
drink_count = 0
items_bought = []



# A tuple consisting of credits that are available. 
accepted_credits = ("25", "50", "100", "500")



# Using dictionaries with a tuple to add two values.
items =  { "A1": ("Candy", 225, "[ Stock: 3 ]"), "A2": ("Sandwich", 400, "[ Stock: 5 ]"),
           "A3": ("Chips", 200, " [ Stock: 1 ]"), "A4": ("Ice Cream", 375, "[ Stock: 3 ]"),
           "B1": ("Water", 100, "[ Stock: 2 ]"), "B2": ("Soda", 225, "[ Stock: 2 ]"),
           "B3": ("Juice", 150, "[ Stock: 1 ]"), "B4": ("Coffee", 325, "[ Stock: 2 ]"),
         }

snacks = { "A1": ("Candy", 225, "[ Stock: 3 ]"), "A2": ("Sandwich", 400, "[ Stock: 5 ]"),
         "A3": ("Chips", 200, " [ Stock: 1 ]"), "A4": ("Ice Cream", 375, "[ Stock: 3 ]")
         }

drinks = {"B1": ("Water", 100, "[ Stock: 2 ]"), "B2": ("Soda", 225, "[ Stock: 2 ]"),
         "B3": ("Juice", 150, "[ Stock: 1 ]"), "B4": ("Coffee", 325, "[ Stock: 2 ]"),
         }



# A dictionary to keep the stock of specific items.
item_stock = {"A1": 3, "A2": 5,
              "A3": 1, "A4": 3,
              "B1": 2, "B2": 2,
              "B3": 1, "B4": 2
              }





# prints the items in an organised manner.
def display():
  print("=======================================================================")
  print("                             [ SNACKS ]")
  for keys, values in snacks.items():
    print("               [" + keys + "] : " + str(values))
  print()

  print("                             [ DRINKS ]")
  for keys, values in drinks.items():
    print("               [" + keys + "] : " + str(values))
  print("=======================================================================\n\n\n")



def money_system():
  global credits, entered_amount, kept_amount

  while True:
    amount = input("Enter amount of accepted credits. Enter \"quit\" to exit:\n")
    print()

    if amount in accepted_credits: # Only accepts credits that are in accepted_credits.

      if int(amount) > credits: # If input is more than the amount of credits, this code will run.
        print("[ You don't have enough credits. ]\n\n")

      else:
        entered_amount += int(amount) # Storing a variable into another variable to keep the information.
        credits -= int(amount) # Deducts the credits by the amount of credits added by user.

        print(f"You added: ({amount})")
        print(f"You now have: ({entered_amount})\n\n")

    elif entered_amount == 0 and amount == "quit": # If no credits are added, this statement prints.
      print("[ No credits entered. ]\n\n") 

    elif amount.lower() == "quit": # Prints and break if user inputs "quit".
      print("[ Entered quit. ]") 
      break

    else:
      print("[ Credit not accepted. ]\n\n")





def item_system():
  global credits, entered_amount, kept_amount, overall_cost, \
  items_bought, snack_count, drink_count
  
  system = True
  while system:

    answer = input("Enter item code. Enter \"quit\" to exit:\n")
    print()
    if answer.upper() in items: # Only accepts inputs that are in items.
      if items[answer.upper()][1] > entered_amount: # If the item cost is more than the amount inputted, this code will run.
        print("[ Amount of money is insufficient. ]\n\n") 

      else:
        print(f"You successfully inputted: {items[answer.upper()][0]}")
        if item_stock[answer.upper()] >= 1: # Checks if item stock is more than 0. Only runs the coomand if item stock has more than 0.
          item_stock[answer.upper()] -= 1 # Deducts the stock by 1.

          entered_amount -= items[answer.upper()][1] # Deducts the amount added by the item cost.
          overall_cost += items[answer.upper()][1] # Adds the amount to the overall cost for the user to see.
          items_bought.append(items[answer.upper()][0]) # Sends the item into a list where the user can see the items bought.

          print(f"Cost of {items[answer.upper()][0]}: ({items[answer.upper()][1]})")
          print(f"You now have: {(entered_amount)}")
          print(f"Amount in stock: {(item_stock[answer.upper()])}\n\n")
          
          if answer in snacks: # Uses an if-in statement to check if the item is either a snack or drink.
           if snack_count == 0:
              print("[ I see you inputted a snack. Would you like a drink with that? ]")
              drink_count += 1
              
          else:
            if drink_count == 0 :
              print("[ I see you inputted a drink. Would you like a snack with that? ]")             
              snack_count += 1
              
        else:
          print("[ That item is out of stock. ]\n\n")  

    elif not items_bought and answer == "quit": # If no items are inputted, this statement prints.
      print("[ No items entered. ]\n\n") 
    
    elif answer == "quit" and snack_count == 1: # If user inputs "quit" and bought only one snack.
      run = True

      while run:
        query = input("Are you sure? Enter \"y\" to quit, enter \"n\" to continue:\n").lower()
        print()

        if query == "y":
          run = False
          system = False

        elif query == "n":
          run = False

          display()

          print(f"[ Your credits: ({credits}) ]")
          print(f"[ Accepted Credits: {accepted_credits}]")
          print(f"[ Amount of credits added: ({entered_amount}) ]\n\n")

          money_system()
          
          print(f"You have added {(entered_amount)} to the machine.\n\n")

          display()

          snack_count += 1

        else:
          print("[ Invalid Input. ]")
          
    elif answer == "quit" and drink_count == 1: # If user inputs "quit" and bought only one drink.
       run = True

       while run:
         query = input("Are you sure? Enter \"y\" to quit, enter \"n\" to continue:\n").lower()
         print()

         if query == "y":
           run = False
           system = False

         elif query == "n":
          run = False

          display()

          print(f"[ Your credits: ({credits}) ]")
          print(f"[ Accepted Credits: {accepted_credits}]")
          print(f"[ Amount of credits added: ({entered_amount}) ]\n\n")

          money_system()

          print(f"You have added {(entered_amount)} to the machine.\n\n")

          display()

          drink_count += 1

         else:
           print("[ Invalid Input. ]")

    elif answer.lower() == "quit": # Prints and break if user inputs "quit".
      print("[ Entered quit. ]\n\n")
      system = False

    else: 
      print("[ That item is not in stock. ]\n\n")





def reciept_system():
  global items_bought, kept_amount, overall_cost, entered_amount

  while True:
    receipt = input("Would you like a reciept? if so, enter \"y\", else enter \"n\":\n").lower() # Asks the user an input regarding the reciept.
    print()

    if receipt == "y": # If the input is "yes", it prints this line.
      print("[You have entered yes. ]\n\n")
      print("Receipt:")
      print("=======================================================================")
      print(f"Items bought: {items_bought}\n")
      print(f"You added: ({kept_amount})")
      print(f"Your overall costs of items: ({overall_cost})")
      print(f"Your change is: ({entered_amount})")
      print("=======================================================================\n\n\n")
      break

    elif receipt == "n": # If the input is "no", it prints this line.
      print("[You have entered no. ]\n\n")
      print("=======================================================================")
      print(f"Items bought: {items_bought}\n")
      print("=======================================================================\n\n\n")
      break

    else: # If the input is neither "yes" or "no", it prints this line.
      print("[ Invalid response, try again. ]")





print("******************[ WELCOME TO THE VENDING MACHINE. ]******************\n")



# prints the items in an organised manner.
display() 


# prints the amount of credits and credits that are allowed.
print(f"[ Your credits: {(credits)}]")
print(f"[ Accepted Credits: {accepted_credits}]\n\n")


# Using a while-loop to loop until the user enters a correct input or if the user enters "quit".
print("[ Please add some credits to the machine. ]")
money_system()

kept_amount = entered_amount # Storing a variable into another variable to keep the information.


print(f"You have added {(entered_amount)} to the machine.\n\n")


# prints the items in an organised manner.
display() 


# Using a while-loop to loop until the user enters a correct input or if the user enters "quit".
print("[ Please input an item code into the machine. ]")
item_system()


credits += entered_amount # Returning the "change" back to the credits.

# Prints the transaction for the user.
reciept_system()


print("*******************[ THANK YOU FOR YOUR PATRONAGE! ]*******************")
```
