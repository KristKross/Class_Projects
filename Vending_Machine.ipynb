```python
# The variables the programme requires.
credits, overall_credits = 550, 550
change, overall_cost = 0, 0
entered_amount, kept_amount = 0, 0
items_bought = []

# A tuple consisting of integers that are available. 
accepted_credits = ("25", "50", "100", "500")

# Using a dictionary with a tuple to add two values.
items = {"A1": ("Water", 100, "Stock: 2"), "A2": ("Soda", 225, "Stock: 2"),
         "A3": ("Juice", 150, "Stock: 1"), "A4": ("Coffee", 325, "Stock: 2"),
         "B1": ("Candy", 225, "Stock: 3"), "B2": ("Sandwich", 400, "Stock: 5"),
         "B3": ("Chips", 200, "Stock: 1"), "B4": ("Ice Cream", 375, "Stock: 3")
         }

# A dictionary to keep the stock of specific items.
item_stock = {"A1": 2, "A2": 2,
              "A3": 1, "A4": 2,
              "B1": 3, "B2": 5,
              "B3": 1, "B4": 3
              }

# prints the items in an organised manner.
print("[ VENDING MACHINE ]\n")
for keys, values in items.items():
    print("[" + keys + "] : " + str(values))
print()


# prints the amount of credits and credits that are allowed.
print("[ Credits: (" + str(credits) + ") ]")
print("[ Accepted Credits: " + str(accepted_credits) + " ]\n")


# Using a while-loop to loop until the user enters a correct input or if the user enters "quit".
while True:

  amount = input("Enter amount of accepted credits. Enter \"quit\" to exit: ")

  if amount in accepted_credits: # Only accepts credits that are in accepted_credits.
    print("[ Credit accepted. ]")
    if int(amount) > credits:
      print("[ You don't have enough credits. ]\n")
    else:
      entered_amount += int(amount) # Using two variables and keeping one of them unchanged.
      credits -= int(amount) # Deducts the credits by the amount of credits added by user.
      print("\nYou added: (" + str(amount) + ")")
      print("[ You now have: (" + str(entered_amount) + ") ]\n")
  elif entered_amount == 0 and amount == "quit": # If no credits are added, this statement prints.
    print("[ No credits entered. ]\n") 
  elif amount.lower() == "quit": # Prints and break if user inputs "quit".
    print("[ Entered quit. ]\n") 
    break
  else:
    print("[ Credit not accepted. ]\n")

kept_amount = entered_amount # Using two variables and keeping one of them unchanged.
print("\nYou have added: (" + str(entered_amount) + ")\n")


for keys, values in items.items():
    print("[" + keys + "] : " + str(values))
print("\n")


# Using a while-loop to loop until the user enters a correct input or if the user enters "quit".
while True:

  answer = input("Enter item code. Enter \"quit\" to exit: ")

  if answer.upper() in items: # Only accepts inputs that are in items.
    if items[answer.upper()][1] > entered_amount: 
      print("[ Amount of money is insufficient. ]\n") 
    else:
      print("\nYou successfully inputted: " + items[answer.upper()][0] + "[Stock: " + str(item_stock[answer.upper()]) + "]\n")
      if item_stock[answer.upper()] >= 1: # Checks if item stock is more than 0. Only runs the coomand if item stock has more than 0.
        item_stock[answer.upper()] -= 1 # Deducts the stock by 1.
        entered_amount -= items[answer.upper()][1] # Deducts the amount added by the item cost.
        overall_cost += items[answer.upper()][1] # Adds the amount to the overall cost for the user to see.
        items_bought.append(items[answer.upper()][0]) # Sends the item into a list where the user can see the items bought.
        print("[ Cost of " + str(items[answer.upper()][0]) + ": (" + str(items[answer.upper()][1]) + ") ]")
        print("[ You now have: (" + str(entered_amount) + ") ]\n")
      else:
        print("[ That item is out of stock. ]\n") # If item stock is 0 this statement prints 
  elif not items_bought and answer == "quit": # If no items are inputted, this statement prints.
    print("[ No items entered. ]\n")    
  elif answer.lower() == "quit": # Prints and break if user inputs "quit".
    print("[ Entered quit. ]\n") 
    break
  elif not answer.upper() in items: 
    print("[ That item is not in stock. ]\n")


change = entered_amount # Equalling the change with entered amount to show amount of change.
credits += change # Returning the change back to the credits 

# Prints the transaction for the user.
print("You bought: " + str(items_bought) + "\n")
print("You added: (" + str(kept_amount) + ")")
print("Your overall costs: (" + str(overall_cost) + ")")
print("Your change is: (" + str(change) + ")\n")
print("Your credits before: (" + str(overall_credits) + ")")
print("Your credits now: (" + str(credits) + ")\n")
print("[ THANK YOU FOR YOUR PATRONAGE! ]")
```
